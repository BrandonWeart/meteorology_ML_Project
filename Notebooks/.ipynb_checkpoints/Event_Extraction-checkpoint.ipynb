```python
import os
import pandas as pd
import numpy as np

# 
# 1) Build set of available dates from your .nc filenames
nc_dir = '/home/scratch/ahaberlie1/gefs_reforecast_preproc/'
files = [f for f in os.listdir(nc_dir) if f.endswith('_daily_preproc_ml.nc')]

available_dates = {
    pd.to_datetime(f.split('_')[0], format='%Y%m%d%H%M').date()
    for f in files
}

# 
# 2) Read & preprocess your tornado data
tor = pd.read_csv(
    "/home/z1995995/data_science_for_the_geosciences/495/1950-2023_actual_tornadoes.csv"
)
tor['date'] = pd.to_datetime(tor['date'], errors='coerce')
tor['day']  = tor['date'].dt.date

date_lower = pd.to_datetime('2001-01-01').date()
date_upper = pd.to_datetime('2019-12-31').date()

tor_filtered = tor[
    (tor['slat'] >= 30) & (tor['slat'] <= 36) &
    (tor['slon'] >= -93) & (tor['slon'] <= -80) &
    (tor['day']  >= date_lower) &
    (tor['day']  <= date_upper)
]

# 
# 3) Group and label, but only if the day is in available_dates
grouped = tor_filtered.groupby('day')
label1_days = []
label2_days = []

for day, g in grouped:
    if day not in available_dates:
        continue
    mags = g['mag']
    if mags.ge(2).any():
        label2_days.append(day)
    elif mags.le(1).all():
        label1_days.append(day)

# 
# 4) Build label0 (no-tornado days), only include available_dates
all_days     = pd.date_range(date_lower, date_upper, freq='D').date
present_days = set(tor_filtered['day'])
label0_days  = [
    d for d in all_days
    if (d not in present_days) and (d in available_dates)
]

# 
# 5) Exclude any hard-coded bad dates
exclusion = {
    pd.to_datetime('2011-04-27').date(),
    pd.to_datetime('2023-03-31').date(),
    pd.to_datetime('1974-04-04').date()
}
for lst in (label0_days, label1_days, label2_days):
    lst[:] = [d for d in lst if d not in exclusion]

# 
# 6) Sort and build DataFrames for all valid dates (no sampling)
sample_label0 = sorted(label0_days)
sample_label1 = sorted(label1_days)
sample_label2 = sorted(label2_days)

def make_df(dates, lbl):
    return pd.DataFrame({
        'date':       [d.strftime('%Y-%m-%d') for d in dates],
        'classification': lbl
    })

df0 = make_df(sample_label0, 0)
df1 = make_df(sample_label1, 1)
df2 = make_df(sample_label2, 2)

# 
# 7) (Optional) concatenate into one DataFrame & export
all_labeled = pd.concat([df0, df1, df2], axis=0).reset_index(drop=True)
all_labeled.to_csv('all_labeled_dates.csv', index=False)

# 
# 8) Inspect
print("No-tornado days:", len(df0))
print("Weak-tornado days:", len(df1))
print("Strong-tornado days:", len(df2))
```

```python
import os
import pandas as pd
import numpy as np

# 
# 1) Build set of available dates from your .nc filenames
nc_dir = '/home/scratch/ahaberlie1/gefs_reforecast_preproc/'
files = [f for f in os.listdir(nc_dir) if f.endswith('_daily_preproc_ml.nc')]

available_dates = {
    pd.to_datetime(f.split('_')[0], format='%Y%m%d%H%M').date()
    for f in files
}

# 
# 2) Read & preprocess your tornado data
tor = pd.read_csv(
    "/home/z1995995/data_science_for_the_geosciences/495/1950-2023_actual_tornadoes.csv"
)
tor['date'] = pd.to_datetime(tor['date'], errors='coerce')
tor['day']  = tor['date'].dt.date

date_lower = pd.to_datetime('2001-01-01').date()
date_upper = pd.to_datetime('2019-12-31').date()

tor_filtered = tor[
    (tor['slat'] >= 30) & (tor['slat'] <= 36) &
    (tor['slon'] >= -93) & (tor['slon'] <= -80) &
    (tor['day']  >= date_lower) &
    (tor['day']  <= date_upper)
]

# 
# 3) Group and label, but only if the day is in available_dates
grouped = tor_filtered.groupby('day')
label1_days = []
label2_days = []

for day, g in grouped:
    if day not in available_dates:
        continue
    mags = g['mag']
    if mags.ge(2).any():
        label2_days.append(day)
    elif mags.le(1).all():
        label1_days.append(day)

# 
# 4) Build label0 (no-tornado days), only include available_dates
all_days     = pd.date_range(date_lower, date_upper, freq='D').date
present_days = set(tor_filtered['day'])
label0_days  = [
    d for d in all_days
    if (d not in present_days) and (d in available_dates)
]

# 
# 5) Exclude any hard-coded bad dates
exclusion = {
    pd.to_datetime('2011-04-27').date(),
    pd.to_datetime('2023-03-31').date(),
    pd.to_datetime('1974-04-04').date()
}
for lst in (label0_days, label1_days, label2_days):
    lst[:] = [d for d in lst if d not in exclusion]

# 
# 6) Sub-sample at fixed intervals:
#       for label0: take every 20th date
#       for label1: take every 2nd date
#       for label2: keep all
sample_label0 = sorted(label0_days)[::20]
sample_label1 = sorted(label1_days)[::2]
sample_label2 = sorted(label2_days)

# 
# 7) Build DataFrames and format dates
def make_df(dates, lbl):
    return pd.DataFrame({
        'date':           [d.strftime('%Y-%m-%d') for d in dates],
        'classification': lbl
    })

df0 = make_df(sample_label0, 0)
df1 = make_df(sample_label1, 1)
df2 = make_df(sample_label2, 2)

# 
# 8) Concatenate and export
all_labeled = pd.concat([df0, df1, df2], axis=0).reset_index(drop=True)
all_labeled.to_csv('all_labeled_dates.csv', index=False)

# 
# 9) Inspect counts
print("No-tornado days sampled:", len(df0))
print("Weak-tornado days sampled:", len(df1))
print("Strong-tornado days sampled:", len(df2))
```

```python
import pandas as pd
import numpy as np

# load your fullylabeled data
df = pd.read_csv('/home/scratch/BWEART2/tor_RF_event_values_all.csv')

# reproducibility
np.random.seed(0)

train_parts = []
val_parts   = []
test_parts  = []

for label in df['classification'].unique():
    # 1) isolate & shuffle this class
    subset = (
        df[df['classification'] == label]
        .sample(frac=1, random_state=0)
        .reset_index(drop=True)
    )
    
    n = len(subset)
    n_train = int(n * 0.70)        # 70%
    n_val   = int(n * 0.10)        # 10%
    # remaining 20% goes to test
    n_test  = n - n_train - n_val

    # 2) slice out
    train_parts.append(subset.iloc[:n_train])
    val_parts.append(subset.iloc[n_train:n_train+n_val])
    test_parts.append(subset.iloc[n_train+n_val: n_train+n_val+n_test])

# 3) recombine all labels
train = pd.concat(train_parts, axis=0).reset_index(drop=True)
val   = pd.concat(val_parts,   axis=0).reset_index(drop=True)
test  = pd.concat(test_parts,  axis=0).reset_index(drop=True)

# 4) sanity check
print("Train  counts:\n", train['classification'].value_counts(), "\n")
print("Val    counts:\n", val  ['classification'].value_counts(), "\n")
print("Test   counts:\n", test ['classification'].value_counts())

# 5) write out
train.to_csv('train.csv', index=False)
val.to_csv('val.csv',   index=False)
test.to_csv('test.csv',  index=False)
```
